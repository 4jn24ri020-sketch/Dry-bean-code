```python
X = df.drop('Class', axis=1)
Y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
print("LR Accuracy:", accuracy_score(Y_test, lr.predict(X_test)))
print("DT Accuracy:", accuracy_score(Y_test, dt.predict(X_test)))
print("RF Accuracy:", accuracy_score(Y_test, rf.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']
accuracy = [0.86558, 0.89166, 0.92361]
plt.figure(figsize=(10,6))
bars = plt.bar(models, accuracy, color=['blue', 'green', 'red'])
plt.title("Model Accuracy Comparison", fontsize=14)
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4),
             ha='center', va='bottom')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
